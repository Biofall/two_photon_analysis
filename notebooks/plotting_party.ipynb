```python
import matplotlib.pyplot as plt # common plotting library
import numpy as np # we will just use this to make some fake data
```

## Making a little dataset to play with

```python
x = np.arange(10)
slope = 2
intercept = 3
noise = np.random.rand(10)

y = slope * x + intercept + noise
y2 = slope**2 * x + intercept + noise
```

## Generating a Basic Plot

Generating plots is all about making things easy to access and readable. When
you make a plot it should have 'handles' for everything you might want to edit,
investigate later, or use again.

### TASK 1 -- Make two plots

We want a plot of of our two datases (x,y) and (x,y2). We want two separate figures.

```python
# the classic way to plot is this and it is kinda bad and problematic.
plt.plot(x, y, label="y")
# now lets say we want to make a new plot something new in this cell, too.
plt.plot(x, y2, label="y2")
# in this case you will see that the plt method doesn't know what you want
# and slams everything into the same plot.
plt.legend() # it does allow you to make a nice legent tho.
```

```python
# the right way to do it is to build the thing up from the components
# -- a figure that holds the 'axes'
# -- and, an set of axes or maybe multiple axes (e.g. in the context where there are multiple panels)
fig, ax = plt.subplots()
ax.plot(x,y, label="y") # the .plt() method is the same, you just now have control of WHERE it is going (onto ax)
```

```python
fig # you can even plot it later. It is now a variable that holds the plot (a big simplification, but practically true)
```

## OK, lets see the real power of plt.subplots()

This function will free you to be super flexible in how you make a plot. It only
requires you to know what you want before you start plotting

### lets make a plot that has two rows and three columns

```python
fig, ax = plt.subplots(nrows=3, ncols=2)
# you might notice that this looks like total shit
```

```python
fig.tight_layout() # now it is pretty
fig
```

```python
# each of these things is its own ax and we will address
# them like they are in an array
ax[0,0].plot(x,y, label="y")
ax[1,1].plot(x,y2, label="y2")

ax[0,0].legend() # an axis has a legend
fig.legend() # so does a figure, but the figure knows about allll our lines
fig
```

```python
my_layout = """
    AAAA
    BBBC
"""

fig, ax = plt.subplot_mosaic(my_layout)
fig.tight_layout()
ax["A"].plot(x,y)
ax["B"].plot(x,y2)
```

```python
def my_plotting_function(x_data, y_data, axis=None):
    
    if axis is None:
        fig, ax = plt.subplots()
    else:
        fig = axis.get_figure()
        ax = axis
        
    ax.plot(x_data,y_data)
    
    return fig, ax
```

```python
my_plotting_function(x,y)
```

```python
my_big_layout = """
    CCBB
    AAAA
"""

fig, ax = plt.subplot_mosaic(my_big_layout)
fig.tight_layout()
```

```python
my_plotting_function(x,y,axis=ax["C"])
```
